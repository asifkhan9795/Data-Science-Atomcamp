```python
sample_column
```

#### test out all the different regex we did in class below one by one. you can comment out the rest

```python
reg = r"[_-]" #will look for _ and -
reg = r"[A-G]" #will look for letters A to G capital
reg = r"[a-z]+" #will look for small a to z and replace consecutive instances ONCE
reg = r"\s" #will look for spaces. \s represents space. s alone will look for the letter s. 
reg = r"[a-z0-9A-Z\s]" #small and capital letters, numbers and spaces 
reg = r"\w" #all digits and numbers. \W will look for the opposite, i.e. everything BUT digits and numbers



sample_column.str.replace(reg, "")
```

```python
reg = r"^[A-Za-z]" #will look for all capital and small letters
reg = r"^[A-Za-z]+" #will look for all capital and small letters BUT will replace all consecutive instances of these characters with ONE replacement (because of + at the end)
reg = r"[0-9]" #will look for all digits
reg = r"[0-9A-Za-z]" #will look for all capital and small letters and digits
reg = r"^[0-9A-Za-z]" #will look for all capital and small letters and digits but at the START of the input string (^ at the start OUTSIDE sq brackets)
reg = r"[0-9]$" #will look for digits but at the END of the input string ($ at the END outside the sq brackets)
reg = r"[^A-Za-z]" #will look for everything EXCEPT capital and small letters
reg = r"[^A-Za-z]+" #will look for everything EXCEPT capital and small letters (because of ^) AND will replace all consecutive instances of these characters with ONE replacement (because of + at the end)
reg = r"[rav]" #will look for r, a and v INDIVIDUALLY
reg = r"(rav)" #will treat rav as a group and look for this exact pattern "rav"



sample_column.str.replace(reg, "")
```

### PD.CUT

This is a way to change numerical, continuous data to categorical data

```python
#creating a small separate dataframe df2 and filling it with the age column from df

df2 = pd.DataFrame(df["Age"].copy(), columns = ["Age"])

df2
```

```python
#here we just gave 5 bins and pd.cut equally divided the ages in the age column into five equal age gaps

pd.cut(df2["Age"], bins = 5) 
```

```python
labels = ["teen", "young adult", "adult", "senior"]
bins = [13, 20, 30, 60, 100]


#adding a new column Age Category to df2 using pd.cut
#the kwarg right = True/False tells us if we want the right (upper) bound to be included. 
#if right = True, then for example 13-20 will be teens INCLUSIVE (i.e. age 20 will also come under teen)
#if right = False, then upper bound isn't included so teen would be less than 20, NOT inclusive of 20
#this applies to all the bins

df2["Age Category"] = pd.cut(df2["Age"], bins = bins, labels = labels, right = False)
```

```python
df2
```

```python
#as you can see, setting right = False results in age 20 being included in young adult

df2.loc[df2["Age"] == 20, :]
# df2.loc[df2["Age"] == 30, :]
```
